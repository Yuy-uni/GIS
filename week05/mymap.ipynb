```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
reitaku = [35.8330, 139.9550]
myhome = [35.8543189,139.9877267]
ArioKashiwa = [35.83346048858035, 140.01288688115926]
SevenEleven=[35.85214336769844, 139.98194702063142]
KashiwaShrine=[35.85989290600351, 139.9737825719323]
# make the map
m = folium.Map(location=reitaku, zoom_start=15)

# add a marker
folium.Marker(reitaku, popup="Reitaku University, the best place on earth!", tooltip="Reitaku University",icon=folium.Icon(color='red')).add_to(m)
folium.Marker(myhome,popup="My home is here, surrouding beautiful nature",tooltip="My home",icon=folium.Icon(color='lightred')).add_to(m)
folium.Marker(ArioKashiwa,popup="The largest mall nearby my home is here, several useful goods are included.",tooltip="ArioKashiwa",icon=folium.Icon(color='green')).add_to(m)
folium.Marker(SevenEleven,popup="The closest convenience store is here, almost daily tools are sold there.",tooltip="SevenEleven",icon=folium.Icon(color='beige')).add_to(m)
folium.Marker(KashiwaShrine,popup="Many festival is held here, they enjoy it as the annual traditional event.",tooltip="Kashiwashrine",icon=folium.Icon(color='purple')).add_to(m)

# show the map
m
```

```python
m.save("yuyamap.html")
```
